```python
N, M = map(int, input().split())
li = list(map(int, input().split()))
cnt = 0
for i in range(N):
    t = 0
    for j in range(i, N):
        t += li[j]
        if t > M:
            break
        elif t == M:
            cnt += 1
print(cnt)
```
